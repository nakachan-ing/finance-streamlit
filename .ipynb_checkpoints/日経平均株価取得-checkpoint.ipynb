```python
# ライブラリの読み込み
import pandas as pd
import time
import yfinance as yf
```

```python
filename = 'nikkei.csv'
df = pd.read_csv(filename, encoding='shift-jis')
df
```

```python
list = []
names = df['銘柄名']
# names[0]

i = 0
for name in names:
    code = df[df['銘柄名']==name]['コード'][i]
    stockCode = str(code) + '.T'
    list.append(stockCode)
    i += 1

    
list

#　取得期間を指定する
days = 15

# 期間を指定して株情報を取得する
histories = []

j = 0
for hist in list[:5]:
    hist_all = yf.Ticker(stockCode).history(period=f'{days}d')
    hist = hist_all[['Close']]
    hist.columns = [df['銘柄名'][j]]
    hist.index = hist.index.strftime('%d %B %Y')
    histories.append(hist)
    j += 1
histories[0]
```
